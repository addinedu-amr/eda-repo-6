```python
import pandas as pd
import folium
import json
```

```python
people = pd.read_csv('./인구밀도_20221115150908.csv')
people
```

```python
people = people.iloc[2:]
```

```python
people
```

```python
people = people[people["동별(3)"].isin(["소계"])]
```

```python
people = people[['동별(2)', '2021.2']]
people.rename(columns={people.columns[0] : "구"}, inplace = True)
people.rename(columns={people.columns[1] : "2021년 인구밀도"}, inplace = True)
people = people.reset_index()
people = people.drop(['index'], axis=1)
people
```

```python
people = people.astype({'2021년 인구밀도':'int'})
```

```python
people = people.set_index('구')
people
```

```python
geo_path = './skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
my_map = folium.Map(location=[37.5502,126.982], zoom_start = 11, tiles='OpenStreetMap')
my_map.choropleth(
    geo_data=geo_str,
    data=people['2021년 인구밀도'],
    columns=[people.index, people['2021년 인구밀도']],
    fill_color='PuRd',
    key_on="feature.id",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="2021년 구별 인구밀도",
)
folium.GeoJson(
    geo_str,
    style_function = lambda x: {'fillColor':'#00000000',
                                'color':'black',
                                'weight' :'2'}
).add_to(my_map)
```
