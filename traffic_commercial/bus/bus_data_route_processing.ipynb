# 원본에서 가공했던 데이터를 다시 가공하는 과정입니다.
### 여기서 원하는 것은 정류장별로 버스노선이 얼마나 되는지 분석하기 위한 과정입니다.
- 2022년 자료를 기준으로 진행합니다.

```python
import pandas as pd
```

```python
bus_2022 = pd.read_csv('./2022_bus_commuter.csv')
bus_2022
```

```python
len(bus_2022)
```

```python
bus_2022[bus_2022['월'].isin([1])]['표준버스정류장ID'].value_counts()
```

# 연도별로 정리되니 자료를 월단위로 구분하기 위해 월을 추가합니다.

```python
bus_2022_list = []
for i in range(1,12):
    bus_2022_list.append(bus_2022[bus_2022['월'].isin([i])])

bus_2022_list[0]
```

# 노선의 개수를 Count 하기 때문에 중복자료를 지우듯이 지웁니다.
## 이전에 value_counts로 정류장별 노선의 개수를 알아내었기 때문에 괜찮습니다

```python
bus_route_2022_1_list = bus_2022_list[9].drop_duplicates(['표준버스정류장ID'])
bus_route_2022_1_list
```

# value_counts를 데이터 프레임으로 만들어버립니다.

```python
bus_route_count = pd.DataFrame(bus_2022[bus_2022['월'].isin([1])]['표준버스정류장ID'].value_counts())
bus_route_count = bus_route_count.reset_index()
bus_route_count.rename(columns={bus_route_count.columns[1]: "환승 가능한 버스의 수"}, inplace=True)
bus_route_count
```

```python
bus_route_count.rename(columns={bus_route_count.columns[0]: "표준버스정류장ID"}, inplace=True)
bus_route_count
```

```python
bus = pd.DataFrame(bus_route_2022_1_list[['역명', '표준버스정류장ID']])
bus
```

# 두 개의 데이터를 merge하면 하나의 자료로 합칠 수 있습니다.

```python
bus_route_count =pd.merge(bus_route_count, bus, left_on='표준버스정류장ID', right_on='표준버스정류장ID', how='inner')
bus_route_count
```

# 이 데이터를 따로 저장하여 마무리합니다.

```python
bus_route_count.to_csv('./2022_10_bus_route.csv', index=False)
```
