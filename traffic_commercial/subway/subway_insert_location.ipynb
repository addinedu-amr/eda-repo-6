# 가공한 데이터에 위치정보를 삽입하는 과정을 보입니다.
- 서울시 지하철 위도경도 데이터를 사용했습니다.

```python
import pandas as pd
```

```python
metro_2022_10 = pd.read_csv('./1차가공/metro_2022.csv')
metro_2022_10
```

```python
metro_location = pd.read_csv('./station_location.csv', encoding="EUC-KR")
metro_location
```

```python
metro_location['lat'].value_counts()
```

```python
metro_check = metro_location[metro_location['lat'].isin([37.554648])]
metro_check
```

```python
len(metro_2022_10)
```

```python
metro_2022_10_file =pd.merge(metro_2022_10, metro_location , left_on='지하철역', right_on='name', how='left')
metro_2022_10_file
```

```python
metro_2022_10_file_2 = metro_2022_10_file.drop(["line", "name", "code"], axis=1)
metro_2022_10_file_2.head(10)
```

```python
metro_2022_10_file_3 = metro_2022_10_file_2.drop_duplicates(['호선명', '지하철역'])
metro_2022_10_file_3
```

# 데이터 수량이 계속 달라서 수량을 맞추기 위해서 전처리를 하다보니.
