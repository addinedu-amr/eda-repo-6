```python
import pandas as pd
```

```python
gage_open_clo = pd.read_csv('./2020_서울시 우리마을가게 상권분석서비스(상권-점포).csv', encoding='EUC-KR')
gage_open_clo
```

```python
gage_open_clo_process = gage_open_clo[gage_open_clo['기준_년_코드'].isin([2020])]
gage_open_clo_process
```

```python
gage_open_clo_process = gage_open_clo_process[gage_open_clo_process['기준_분기_코드'].isin([2])]
gage_open_clo_process
```

# 업종별을 날리는거라 아쉽긴 해도..

```python
gage_open_clo_process = gage_open_clo_process.drop(['상권_구분_코드', '상권_코드_명', '서비스_업종_코드', '서비스_업종_코드_명'], axis=1)
```

```python
gage_open_clo_process
```

```python
import numpy as np
```

```python
gage_open_clo_process_pivot = gage_open_clo_process.pivot_table(gage_open_clo_process, index=["상권_코드"], aggfunc = [np.sum] )
gage_open_clo_process_pivot
```

```python
gage_open_clo_process_pivot.columns = ['/'.join(col) for col in gage_open_clo_process_pivot.columns]
gage_open_clo_process_pivot
```

```python
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[0]: "개업_율"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[1]: "개업_점포_수"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[2]: "기준_년_코드"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[3]: "기준_분기_코드"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[4]: "유사_업종_점포_수"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[5]: "점포_수"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[6]: "폐업_률"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[7]: "폐업_점포_수"}, inplace=True)
gage_open_clo_process_pivot.rename(columns={gage_open_clo_process_pivot.columns[8]: "프랜차이즈_점포_수"}, inplace=True)
```

```python
gage_open_clo_process_pivot.head(20)
```

```python
gage_open_clo_process_pivot = gage_open_clo_process_pivot.drop(['개업_율', '폐업_률'], axis=1)
gage_open_clo_process_pivot
```

```python
gage_open_clo_process_pivot = gage_open_clo_process_pivot.reset_index()
```

```python
gage_open_clo_process_pivot["연도_분기"] = "2020_2분기"
gage_open_clo_process_pivot
```

```python
gage_open_clo_process_pivot.columns
```

```python
gage_open_clo_process_pivot = gage_open_clo_process_pivot[['연도_분기', '상권_코드', '개업_점포_수', '기준_년_코드', '기준_분기_코드', '유사_업종_점포_수', '점포_수',
       '폐업_점포_수', '프랜차이즈_점포_수']]
gage_open_clo_process_pivot
```

```python
gage_open_clo_process_pivot.to_csv('./2020_상권별 개점_폐업.csv', encoding='EUC-KR', index=False)
```
