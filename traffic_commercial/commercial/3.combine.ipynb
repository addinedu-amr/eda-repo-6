```python
location = pd.read_csv('./서울시 상업지역 위치정보_가공.csv', encoding='EUC-KR')
location
```

```python
location["long"] = 0.
location["lat"] = 0.
location.info()
```

```python
a = Proj(init = 'epsg:5181')
b = Proj(init = 'epsg:4326')

for i in range(len(location)):
    x1, y1 = location['엑스좌표_값'][i], location['와이좌표_값'][i]
    x2, y2 = transform(a, b, x1, y1)
    locaiton["long"][i] = x2
    locaiton["lat"][i] = y2

locaiton
```

```python
locaiton = locaiton.drop(['엑스좌표_값', '와이좌표_값'], axis=1)
locaiton
```

```python
locaiton.to_csv('./상권 위치좌표_최종.csv', encoding='EUC-KR', index=False)
```

```python
sale = pd.read_csv('./2019_상권별_매출_1차가공.csv', encoding='EUC-KR')
sale
```

```python
sale_location = pd.merge(sale, locaiton, left_on='상권_코드', right_on='상권_코드', how='left')
sale_location
```
