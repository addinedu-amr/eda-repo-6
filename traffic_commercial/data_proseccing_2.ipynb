```python
import pandas as pd
```

```python
bus_2019 = pd.read_csv('./2019_bus_commute.csv')
bus_2019 = bus_2019.drop('노선ID', axis=1)
bus_2019
```

```python
bus_2019 = bus_2019.astype({'사용년월' : 'string'})
bus_2019['월'] = bus_2019['사용년월'].str[4:]
bus_2019 = bus_2019.drop('사용년월', axis = 1 )
bus_2019 = bus_2019[['역명', '연도', '월', '총 승차 인원', '총 하차 인원', '출근시간(04~09) 승차', '출근시간(07~10) 하차',
                    '퇴근시간(16~20) 승차', '퇴근시간(17~21) 하차', '늦은시간(20~00) 승차', '늦은시간(21~02) 하차', '점심시간', '저녁시간']]
bus_2019
```

```python
bus_2019.to_csv('./2019_bus_commuter', index=False)
```
