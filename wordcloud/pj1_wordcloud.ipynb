```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
with open('./WordCloudtext', 'r', encoding='utf-8') as f:
    text = f.read()

okt = Okt()
nouns = okt.nouns(text)

words = [n for n in nouns if len(n) > 1]

c = Counter(words)
```

```python
img = Image.open('seoul.png')
img_array = np.array(img)

wc = WordCloud(font_path='NanumBarunGothic.ttf', background_color='white', colormap= 'Accent_r',
                 width=400, height=400, scale=2.0, max_font_size=200, mask=img_array)
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
```

```python
wc.to_file('pj1.png')
```
